```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

SHAPE_FILE = '/course/food_access/tl_2010_53_tract00/tl_2010_53_tract00.shp'
FOOD_ACCESS_FILE = '/course/food_access/food_access.csv'
```

```python
census = gpd.read_file(SHAPE_FILE)
food_access = pd.read_csv(FOOD_ACCESS_FILE)
merged = census.merge(food_access, left_on='CTIDFP00', right_on='CensusTract', how='left')
merged = gpd.GeoDataFrame(merged, geometry='geometry')  # make it into GeoDataFrame
```
